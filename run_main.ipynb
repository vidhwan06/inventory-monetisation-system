# Run the inventory-monetisation-system main program

This notebook imports required libraries, sets up the environment, runs `main.py`, and displays the captured output.

```python
# Import Required Libraries
import os
import subprocess
import sys
```

```python
# Set Up Environment Variables
os.environ.setdefault('PYTHONUNBUFFERED', '1')
# Add any additional environment variables here if needed
```

```python
# Run the Main Program
program_path = '/workspaces/inventory-monetisation-system/main.py'
result = subprocess.run([sys.executable, program_path], cwd=os.path.dirname(program_path), capture_output=True, text=True, env=os.environ)
```

```python
# Display Program Output
print('Return code:', result.returncode)
print('--- STDOUT ---')
print(result.stdout)
print('--- STDERR ---')
print(result.stderr)
```
